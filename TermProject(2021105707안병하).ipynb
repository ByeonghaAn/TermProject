```python
import time
import urllib.request
from pytube import Playlist
from pytube import YouTube
from pymongo import MongoClient

my_client = MongoClient("mongodb://localhost:27017/")

mydb = my_client['Tproject']
mycol = mydb['Youtube']

playlist = Playlist('https://www.youtube.com/playlist?list=PLmtapKaZsgZt3g_uAPJbsMWdkVsznn_2R')
```

```python
for i in range(0,len(playlist)//4):
    url = playlist[i]
    res = urllib.request.urlopen(url)
    if(res.status == 200):
        yt = YouTube(url)
        mycol.insert_one({"id":i, "title":yt.title, "thumbnail":yt.thumbnail_url, "author":yt.author, "views":yt.views, "keywords":yt.keywords, "url":url})
        print(i)
        time.sleep(8)
    if(i%10 == 0):
        time.sleep(30)
```
